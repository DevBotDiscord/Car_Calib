# Calibration Evaluation Quickstart
Evaluate recorded frames through the shared calibration path and inspect review artifacts.

```python
from pathlib import Path
import sys

# VS Code/Jupyter commonly starts this notebook inside notebooks/. Find the
# repository root explicitly so local project modules are importable.
candidates = [Path.cwd(), *Path.cwd().parents]
REPO_ROOT = next(
    (path for path in candidates if (path / "calibration_evaluation.py").exists()),
    None,
)
if REPO_ROOT is None:
    raise RuntimeError("Repository root not found. Open this notebook from the experiment worktree.")
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from calibration_evaluation import CalibrationEvaluator, iter_video_frames

# Change this to a real recorded video on your machine.
VIDEO_PATH = REPO_ROOT / "videos" / "route.mp4"
OUTPUT_DIR = REPO_ROOT / "evaluations" / "notebook_run"

print(f"Repository: {REPO_ROOT}")
print(f"Python: {sys.executable}")
print(f"Video: {VIDEO_PATH}")
```

```python
if not VIDEO_PATH.exists():
    raise FileNotFoundError(
        f"Video not found: {VIDEO_PATH}\n"
        "Set VIDEO_PATH in the first cell to an existing .mp4/.avi/.mov/.mkv file."
    )

evaluator = CalibrationEvaluator(output_dir=OUTPUT_DIR, review_every=30)
try:
    records, summary = evaluator.evaluate_frames(
        iter_video_frames(VIDEO_PATH, max_frames=300)
    )
finally:
    evaluator.close()
summary.as_dict()
```
